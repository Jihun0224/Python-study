```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
point = np.array([[-2.9, 35.4],
                 [-2.1, 19.7],
                 [-0.9, 5.7],
                 [0.1, 1.2],
                  [1.1, 2.1],
                 [1.9, 8.7],
                 [3.1, 25.7],
                 [4.0, 41.5]])
```

```python

for i in range(8) : 
    point = np.array([[-2.9, 35.4],
                 [-2.1, 19.7],
                 [-0.9, 5.7],
                 [0.1, 1.2],
                  [1.1, 2.1],
                 [1.9, 8.7],
                 [3.1, 25.7],
                 [4.0, 41.5]])
    print('[{}]을 뺐을 때입니다.'.format(point[i]))
    point= np.delete(point,[2*i,2*i+1]).reshape(-1,2)
    
    x = point[:,0]
    print('x의 좌표 : ', x)
    y = point[:,1]
    print('y의 좌표 : ',y)
    a, b, c = np.polyfit(x, y, 2)
    print('a : {}, b : {}, c: {}'.format(round(a,4), round(b,4), round(c,4)))
    plt.scatter(x,y, c = 'black', s= 100)
    plt.plot(x,y)
#  fit_x = np.arange(-3.4,0.1,0.1)
#  fit_y = (a*fit_x)**2 + b*fit_x + c
#  plt.plot(fit_x, fit_y)
    plt.show()
```
