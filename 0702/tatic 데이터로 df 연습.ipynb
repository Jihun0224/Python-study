# 2번  
2. age의 결측치를 생존여부, 성별에 따라 	grouping 하고 각각의 평균을 결측치에 대입하라

```python
df = pd.read_csv('../data/train.csv')
df1 = df[(df["Survived"] == 1) & (df["Sex"] == "male")]
df2 = df[(df["Survived"] == 0) & (df["Sex"] == "male")]
df3 = df[(df["Survived"] == 1) & (df["Sex"] == "female")]
df4 = df[(df["Survived"] == 0) & (df["Sex"] == "female")]

mean0 = df1['Age'].mean()
mean1 = df2['Age'].mean()
mean2 = df3['Age'].mean()
mean3 = df4['Age'].mean()
```

```python
index = 0
for row in df["PassengerId"]-1:
    if df["Age"].isna()[index]:
        if (df["Survived"][index] == 1) & (df["Sex"][index] == "male"):
            df["Age"][index]=mean0
        if (df["Survived"][index] == 0) & (df["Sex"][index] == "male"):
            df["Age"][index]=mean1
        if (df["Survived"][index] == 1) & (df["Sex"][index] == "female"):
            df["Age"][index]=mean2
        if (df["Survived"][index] == 0) & (df["Sex"][index] == "female"):
            df["Age"][index]=mean3
    index += 1
df["Age"].isna().sum()
```

```python
df1['Age'] = df1['Age'].fillna(mean0)
df2['Age'] = df2['Age'].fillna(mean1)
df3['Age'] = df3['Age'].fillna(mean2)
df4['Age'] = df4['Age'].fillna(mean3)

df_res = df1.append(df2)
df_res = df_res.append(df3)
df_res = df_res.append(df4)
```
